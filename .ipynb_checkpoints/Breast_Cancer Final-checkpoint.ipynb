```python
import numpy as np
import pandas as pd
import pickle
import joblib
import json
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
#Loading the dataset
data_f = pd.read_csv('data.csv')
```

```python
#Viewing the dataset and it's information
print('*' * 100)
print('The first 5 rows of th dataset:')
display(data_f.head(5))
print('*' * 100)
print('Shape of th dataset:')
display(data_f.shape)
print('*' * 105)
print('Columns of the dataset:')
display(data_f.columns)
print('*' * 105)
print('Information of the dataset:')
display(data_f.info())
print('*' * 105)
print('Details of the dataset:')
display(data_f.describe())
```

```python
#Making a copy of the dataset so as not to making changes to the original dataset
dataset = data_f.copy()
```

#### working on categorical data

```python
dataset['diagnosis'] = (dataset['diagnosis'] == 'B').astype(int)
```

#### Splitting the dataset into traning, validation and testing datasets

```python
#splitting the data into trainning, validation and testing
train, val, test = np.split(dataset.sample(frac = 1), [int(0.6 * len(dataset)), int(0.8 * len(dataset))]) 
```

```python
print('Shape of the training set:')
display(train.shape)
print('Shape of the validation set:')
display(val.shape)
print('Shape of the test set:')
display(test.shape)
```

```python
#Splitting the dataset into x and y
y_train = train['diagnosis']
train_input = train.drop(['id', 'diagnosis'], axis = 1)
y_val = val['diagnosis']
val_input = val.drop(['id', 'diagnosis'], axis = 1)
y_test = test['diagnosis']
test_input = test.drop(['id', 'diagnosis'], axis = 1)
```

#### Scaling numeric data

```python
scaler = StandardScaler()
```

```python
#Scaling numric data using StandardScaler
scaler.fit(train_input)
x_train = scaler.transform(train_input)
x_val = scaler.transform(val_input)
x_test = scaler.transform(test_input)
```

#### using a pipeline

```python
#Pipeline of the scaler and Logistic Regression as the model
pipelinelr = Pipeline(steps = [
    ('model', LogisticRegression(solver = 'liblinear'),
    )
   
])
```

```python
#Fitting the pipeline
pipeline.fit(train_input, y_train)
```

```python
#Print the classification report
y_pred = pipeline.predict(val_input)
print(classification_report(y_pred, y_val))
```

```python
pipelinelr = Pipeline(steps = [
    ('feature_selector', SelectKBest(score_func = f_classif, k =10)),
    ('model', LogisticRegression(solver = 'liblinear'),
    )
   
])
```

```python
para_grimlr = {
    'feature_selector__k': [3,8, 10, 12, 14, 16, 18, 20, 30]}
```

```python
#Creating a grid search variable
grid_searchlr = GridSearchCV(
    estimator = pipelinelr,
    param_grid = para_grimlr,
    cv = 10,
    scoring = 'recall',
    n_jobs = -1,
    refit = True,
    return_train_score = True)
```

```python
#Fitting the grid search
grid_searchlr.fit(train_input, y_train)
```

```python
#Printing the best parameters and the best score for the grid search
print(f"Best Parameters:    {grid_searchlr.best_params_}")
print(f"Best CV SCORE:      {grid_searchlr.best_score_:,.2f}")
```

```python
#Getting the classification report of the grid search
best_modellr = grid_searchlr.best_estimator_
y_predict_bestlr = best_modellr.predict(val_input)
val_scorelr = classification_report(y_val, y_predict_best)
```

#### Using GridsearchCV

```python
#Grid search of with Decision tree as the model
pipelinedt = Pipeline(steps = [
    #('scaler', StandardScaler()),
    ('feature_selector', SelectKBest(score_func = f_classif, k =10)),
    ('model', DecisionTreeClassifier(max_depth = 3, max_leaf_nodes = 200 ))
    
])
```

```python
#Parameters for the Grid search
para_grimdt = {
    'feature_selector__k': [3,8, 10, 12, 14, 16, 18, 20, 30],
    'model__max_depth': [3, 50, 80, 100, 150, 200],
    'model__max_leaf_nodes': [3, 50, 80, 100,]
}
```

```python
#Creating a grid search variable
grid_searchdt = GridSearchCV(
    estimator = pipelinedt,
    param_grid = para_grimdt,
    cv = 10,
    scoring = 'recall',
    n_jobs = -1,
    refit = True,
    return_train_score = True)
```

```python
#Fitting the grid search
grid_searchdt.fit(train_input, y_train)
```

```python
#Printing the best parameters and the best score for the grid search
print(f"Best Parameters:    {grid_searchdt.best_params_}")
print(f"Best CV SCORE:      {grid_searchdt.best_score_:,.2f}")
```

```python
#Getting the classification report of the grid search
best_modeldt = grid_searchdt.best_estimator_
y_predict_bestdt = best_modeldt.predict(val_input)
val_scoredt = classification_report(y_val, y_predict_best)
```

```python
print(f"THE VAL SCORE IS:  ")
print(f"{val_scoredt}")
```

```python
#Printing the top 5 model perfoermers
result_df = pd.DataFrame(grid_searchdt.cv_results_)
top5 = result_df.sort_values('rank_test_score').head(5)

print("\nTop 5 Hyperparameter Combinations:")
print(top5[[
    'rank_test_score',
    'param_model__max_depth',
    'param_model__max_leaf_nodes',
    'param_feature_selector__k',
    #'param_preprocessor__num__imputer__strategy',
    #'mean_test_score',
    #'std_test_score',
    #'mean_fit_time'
]].to_string())
```

```python
#Printing the most important features for prediction
pipe = grid_searchdt.best_estimator_
selector = pipe.named_steps['feature_selector']
mask = selector.get_support()
selected_features = np.array(train_input.columns)[mask]
print(selected_features)

#selector = grid_search.best_estimator_.named_steps["feature_selector"]

#selected_features = X_train.columns[selector.get_support()]
#print(selected_features)
```

### RandomForest

```python
#Grid Search using Random Forest Regressor
pipline = Pipeline(steps =[
    ('feature_selectors', SelectKBest(score_func = f_classif, k =10)),
    ('rf_model', RandomForestClassifier(bootstrap=True, min_samples_split =2, n_estimators = 100, max_depth = 3, max_leaf_nodes = 16))
     ])
```

```python
para_grimm = {
    'feature_selectors__k': [3,8, 10, 12, 14, 16, 18, 20, 30],
    'rf_model__n_estimators': [10, 20, 30],
    'rf_model__max_depth': [3, 5, 10, 15, 20],
    'rf_model__max_leaf_nodes': [10, 20, 30],
    'rf_model__min_samples_split': [3, 5,10, 15]
    
}
```

```python
grid_searchcv = GridSearchCV(
    estimator = pipline,
    param_grid = para_grimm,
    cv = 10,
    scoring = 'recall',
    n_jobs = -1,
    refit = True,
    return_train_score = True
)
```

```python
#fitting the grid search for the random forest
grid_searchcv.fit(train_input, y_train)
```

```python
#printing the best paramaters and the cv score
print(f"Best Parameters: {grid_searchcv.best_params_}")
print(f"Best CV Score: {grid_searchcv.best_score_:,.2f}")
```

```python
#Getting the classification report of the grid search
best_model = grid_searchcv.best_estimator_
y_predict_best1 = best_model.predict(val_input)
val_score1 = classification_report(y_val, y_predict_best1)
```
